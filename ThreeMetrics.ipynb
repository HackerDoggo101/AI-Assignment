```python
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords


# Combine text from all rows into a single string
all_text = ' '.join(sampled_df['text'])

# Tokenize the text
tokens = word_tokenize(all_text)

# Remove punctuation and convert to lowercase
tokens = [word.lower() for word in tokens if word.isalnum()]

# Remove stopwords
stop_words = set(stopwords.words('english'))
tokens = [word for word in tokens if word not in stop_words]

# Calculate word frequencies
word_freq = Counter(tokens)

# Display the top 10 most frequent words
top_10_words = word_freq.most_common(10)
print("Top 10 Most Occurring Words:")
for word, frequency in top_10_words:
    print(f"{word}: {frequency}")

# Plot word frequency distribution as a bar graph
plt.figure(figsize=(12, 6))
plt.bar(*zip(*word_freq.most_common(10)),color='orange')
plt.title('Top 10 Most Occurring Words in Lyrics')
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

```

# Text Cleaning

```python
import re
import nltk
from nltk.corpus import stopwords
def cleaning(text):
    text = re.sub("[^a-zA-Z]", " ", text)
    text = text.lower()
    text = text.split()
    stops = set(stopwords.words("english"))
    text = [w for w in text if not w in stops]
    text = " ".join(text)
    return text
```

```python
sampled_df["combined_features"] = sampled_df["combined_features"].apply(func = cleaning)
```

# Word2Vec (Google’s Word2vec Pretrained Word Embedding)

```python
#%pip install gensim
from gensim.models import Word2Vec, KeyedVectors
import gensim.downloader as api
from sklearn.metrics.pairwise import cosine_similarity

# Load the pre-trained Word2Vec model (Google News)
model = api.load('word2vec-google-news-300')

def average_word_vectors(words, model, vocabulary, num_features):
    feature_vector = np.zeros((num_features,), dtype="float64")
    nwords = 0.
    
    for word in words:
        if word in vocabulary: 
            nwords = nwords + 1.
            feature_vector = np.add(feature_vector, model[word])
    
    if nwords:
        feature_vector = np.divide(feature_vector, nwords)
        
    return feature_vector

# Function to compute average word vectors for all songs
def averaged_word_vectorizer(corpus, model, num_features):
    vocabulary = set(model.index_to_key)
    features = [average_word_vectors(tokenized_sentence, model, vocabulary, num_features) for tokenized_sentence in corpus]
    return np.array(features)

# Compute average word vectors for all songs
w2v_feature_array = averaged_word_vectorizer(corpus=sampled_df['combined_features'], model=model, num_features=300)
```

```python
from scipy.stats import pearsonr
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def get_pearson_similarity(user_song_vector, w2v_feature_array):
    # Compute Pearson correlation coefficient between the user song vector and all other song vectors
    similarity_scores = [pearsonr(user_song_vector.flatten(), song_vector)[0] for song_vector in w2v_feature_array]

    return similarity_scores

def get_jaccard_similarity(user_song_text, other_song_text):
    # Convert song texts to sets of words
    user_song_set = set(user_song_text.split())
    other_song_set = set(other_song_text.split())
    
    # Compute Jaccard similarity
    intersection = len(user_song_set.intersection(other_song_set))
    union = len(user_song_set.union(other_song_set))
    jaccard_similarity = intersection / union if union != 0 else 0
    
    return jaccard_similarity

def get_similar_songs(user_song, user_artist, sampled_df, w2v_feature_array):
    # Find the index of the user song and artist
    song_indices = sampled_df[(sampled_df['song'] == user_song) & (sampled_df['artist'] == user_artist)].index
    if len(song_indices) == 0:
        print("Song not found. Please try again.")
        return
    else:
        song_index = song_indices[0]

    # Compute the cosine similarities between the user song and all other songs
    user_song_vector = w2v_feature_array[song_index].reshape(1, -1)
    cosine_scores = cosine_similarity(user_song_vector, w2v_feature_array).flatten()

    # Compute the Jaccard similarities between the user song and all other songs
    jaccard_scores = [get_jaccard_similarity(sampled_df['combined_features'][song_index], other_song_text) for other_song_text in sampled_df['combined_features']]

    # Compute the Pearson similarities between the user song and all other songs
    pearson_scores = get_pearson_similarity(user_song_vector, w2v_feature_array)

    # Get the top 10 most similar songs for each metric
    sorted_cosine_similar_songs = get_top_similar_songs(cosine_scores, sampled_df, 10)
    sorted_jaccard_similar_songs = get_top_similar_songs(jaccard_scores, sampled_df, 10)
    sorted_pearson_similar_songs = get_top_similar_songs(pearson_scores, sampled_df, 10)

    # Print the top 10 similar songs with their similarity scores for each metric
    print("\nSimilar songs using Cosine Similarity:")
    print_top_similar_songs(sorted_cosine_similar_songs, sampled_df, "Cosine Similarity")
    
    print("\nSimilar songs using Jaccard Similarity:")
    print_top_similar_songs(sorted_jaccard_similar_songs, sampled_df, "Jaccard Similarity")
    
    print("\nSimilar songs using Pearson Correlation Coefficient:")
    print_top_similar_songs(sorted_pearson_similar_songs, sampled_df, "Pearsone Correlation Coefficient Similarity")

def get_top_similar_songs(scores, sampled_df, top_n):
    similar_songs = list(enumerate(scores))
    sorted_similar_songs = sorted(similar_songs, key=lambda x: x[1], reverse=True)[1:top_n+1]
    return sorted_similar_songs


def print_top_similar_songs(sorted_similar_songs, sampled_df, metric_name):
    print("Top 10 similar songs using {}:".format(metric_name))
    print("{:<5} | {:<40} | {:<30} | {:<20}".format("Rank", "Song", "Artists", "Similarity Score"))
    print("-" * 90)
    for rank, (i, score) in enumerate(sorted_similar_songs, start=1):
        song_name = sampled_df.loc[i, 'song']
        artists = sampled_df.loc[i, 'artist']
        print("{:<5} | {:<40} | {:<30} | {:.4f}".format(rank, song_name, artists, score))

# Assuming you have loaded your data and preprocessed it
# Here you can add the code to load data and preprocess it


```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from scipy.stats import pearsonr

# Load GloVe embeddings
def load_glove_model(glove_file):
    print("Loading GloVe model...")
    with open(glove_file, 'r', encoding='utf-8') as f:
        glove_map = {}
        for line in f:
            line = line.strip().split()
            word = line[0]
            vec = np.array(line[1:], dtype=np.float64)
            glove_map[word] = vec
    print("Done.")
    return glove_map

glove_file = "glove.6B.100d.txt"
glove_map = load_glove_model(glove_file)

# Function to average word vectors for a text
def average_word_vectors(words, glove_map, num_features):
    feature_vector = np.zeros((num_features,), dtype="float64")
    nwords = 0.
    
    for word in words:
        if word in glove_map:
            nwords = nwords + 1.
            feature_vector = np.add(feature_vector, glove_map[word])
    
    if nwords:
        feature_vector = np.divide(feature_vector, nwords)
        
    return feature_vector

# Function to compute average word vectors for all songs
def averaged_word_vectorizer(corpus, glove_map, num_features):
    features = [average_word_vectors(tokenized_sentence, glove_map, num_features) for tokenized_sentence in corpus]
    return np.array(features)

# Compute average word vectors for all songs
glove_feature_array = averaged_word_vectorizer(corpus=sampled_df['combined_features'], glove_map=glove_map, num_features=100)

# Function to compute Jaccard similarity between two texts
def jaccard_similarity(text1, text2):
    set1 = set(text1.split())
    set2 = set(text2.split())
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union if union != 0 else 0

# Function to compute Pearson correlation coefficient between two vectors
def pearson_similarity(vector1, vector2):
    return pearsonr(vector1, vector2)[0]

# Function to get similar songs using different metrics
def get_similar_songs(user_song, user_artist, sampled_df, glove_feature_array):
    # Find the index of the user song and artist
    song_indices = sampled_df[(sampled_df['song'] == user_song) & (sampled_df['artist'] == user_artist)].index
    if len(song_indices) == 0:
        print("Song not found. Please try again.")
        return
    else:
        song_index = song_indices[0]

    # Compute the user song vector
    user_song_vector = glove_feature_array[song_index]

    # Compute similarity scores using different metrics
    cosine_scores = cosine_similarity([user_song_vector], glove_feature_array).flatten()
    jaccard_scores = [jaccard_similarity(sampled_df['combined_features'][song_index], other_song_text) for other_song_text in sampled_df['combined_features']]
    pearson_scores = [pearson_similarity(user_song_vector, song_vector) for song_vector in glove_feature_array]

    # Get the top 10 most similar songs for each metric
    sorted_cosine_similar_songs = get_top_similar_songs(cosine_scores, sampled_df, 10)
    sorted_jaccard_similar_songs = get_top_similar_songs(jaccard_scores, sampled_df, 10)
    sorted_pearson_similar_songs = get_top_similar_songs(pearson_scores, sampled_df, 10)

    # Print the top 10 similar songs with their similarity scores for each metric
    print("\nSimilar songs using Cosine Similarity:")
    print_top_similar_songs(sorted_cosine_similar_songs, sampled_df, "Cosine Similarity")
    
    print("\nSimilar songs using Jaccard Similarity:")
    print_top_similar_songs(sorted_jaccard_similar_songs, sampled_df, "Jaccard Similarity")
    
    print("\nSimilar songs using Pearson Correlation Coefficient:")
    print_top_similar_songs(sorted_pearson_similar_songs, sampled_df, "Pearson Correlation Coefficient")

def get_top_similar_songs(scores, sampled_df, top_n):
    similar_songs = list(enumerate(scores))
    sorted_similar_songs = sorted(similar_songs, key=lambda x: x[1], reverse=True)[1:top_n+1]
    return sorted_similar_songs


def print_top_similar_songs(sorted_similar_songs, sampled_df, metric_name):
    print("Top 10 similar songs using {}:".format(metric_name))
    print("{:<5} | {:<40} | {:<30} | {:<20}".format("Rank", "Song", "Artists", "Similarity Score"))
    print("-" * 90)
    for rank, (i, score) in enumerate(sorted_similar_songs, start=1):
        song_name = sampled_df.loc[i, 'song']
        artists = sampled_df.loc[i, 'artist']
        print("{:<5} | {:<40} | {:<30} | {:.4f}".format(rank, song_name, artists, score))

# Assuming you have loaded your data and preprocessed it
# Here you can add the code to load data and preprocess it

# Get user input for song name and artist name
user_song = input("Enter a song name: ")
user_artist = input("Enter the artist name: ")

# Get similar songs using all three similarity metrics
get_similar_songs(user_song, user_artist, sampled_df, glove_feature_array)

# # Find the index of the user song
# song_indices = sampled_df[(sampled_df['song'] == user_song) & (sampled_df['artist'] == user_artist)].index
# if len(song_indices) == 0:
#     print("Song not found. Please try again.")
# else:
#     print("Matching song(s) found:")
#     for index in song_indices:
#         song_name = sampled_df.loc[index, 'song']
#         artist_name = sampled_df.loc[index, 'artist']
#         print("Song: {}, Artist: {}".format(song_name, artist_name))

import pickle
pickle.dump(glove_feature_array,open('glove_feature_array.pkl','wb'))
pickle.dump(sampled_df,open('sampled_df.pkl','wb'))

```

# Doc2Vec

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from scipy.stats import pearsonr

# Tag documents with unique identifiers
tagged_data = [TaggedDocument(words=doc.split(), tags=[str(i)]) for i, doc in enumerate(sampled_df['combined_features'])]

# Initialize Doc2Vec model
doc2vec_model = Doc2Vec(vector_size=50, window=5, min_count=1, workers=4, epochs=10)

# Build vocabulary
doc2vec_model.build_vocab(tagged_data)

# Train Doc2Vec model
doc2vec_model.train(tagged_data, total_examples=doc2vec_model.corpus_count, epochs=doc2vec_model.epochs)

# Function to compute document embeddings
def get_document_vector(model, text):
    return model.infer_vector(text.split())

# Compute document embeddings for all songs
doc_embeddings = np.array([get_document_vector(doc2vec_model, text) for text in sampled_df['combined_features']])

# Function to compute cosine similarity between two vectors
def cosine_similarity_vectorized(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))

# Function to compute Jaccard similarity between two texts
def jaccard_similarity(text1, text2):
    set1 = set(text1.split())
    set2 = set(text2.split())
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union if union != 0 else 0

# Function to compute Pearson correlation coefficient between two vectors
def pearson_similarity(vector1, vector2):
    return pearsonr(vector1, vector2)[0]

# Function to get similar songs using different metrics
def get_similar_songs(user_song, user_artist, sampled_df, doc_embeddings):
    # Find the index of the user song and artist
    song_indices = sampled_df[(sampled_df['song'] == user_song) & (sampled_df['artist'] == user_artist)].index
    if len(song_indices) == 0:
        print("Song not found. Please try again.")
        return
    else:
        song_index = song_indices[0]

    # Compute the user song vector
    user_song_vector = doc_embeddings[song_index]

    # Compute similarity scores using different metrics
    cosine_scores = [cosine_similarity_vectorized(user_song_vector, song_vector) for song_vector in doc_embeddings]
    jaccard_scores = [jaccard_similarity(sampled_df['combined_features'][song_index], other_song_text) for other_song_text in sampled_df['combined_features']]
    pearson_scores = [pearson_similarity(user_song_vector, song_vector) for song_vector in doc_embeddings]

    # Get the top 10 most similar songs for each metric
    sorted_cosine_similar_songs = get_top_similar_songs(cosine_scores, sampled_df, 10)
    sorted_jaccard_similar_songs = get_top_similar_songs(jaccard_scores, sampled_df, 10)
    sorted_pearson_similar_songs = get_top_similar_songs(pearson_scores, sampled_df, 10)

    # Print the top 10 similar songs with their similarity scores for each metric
    print("\nSimilar songs using Cosine Similarity:")
    print_top_similar_songs(sorted_cosine_similar_songs, sampled_df, "Cosine Similarity")
    
    print("\nSimilar songs using Jaccard Similarity:")
    print_top_similar_songs(sorted_jaccard_similar_songs, sampled_df, "Jaccard Similarity")
    
    print("\nSimilar songs using Pearson Correlation Coefficient:")
    print_top_similar_songs(sorted_pearson_similar_songs, sampled_df, "Pearson Correlation Coefficient")

def get_top_similar_songs(scores, sampled_df, top_n):
    similar_songs = list(enumerate(scores))
    sorted_similar_songs = sorted(similar_songs, key=lambda x: x[1], reverse=True)[1:top_n+1]
    return sorted_similar_songs

def print_top_similar_songs(sorted_similar_songs, sampled_df, metric_name):
    print("Top 10 similar songs using {}:".format(metric_name))
    print("{:<5} | {:<40} | {:<30} | {:<20}".format("Rank", "Song", "Artists", "Similarity Score"))
    print("-" * 90)
    for rank, (i, score) in enumerate(sorted_similar_songs, start=1):
        song_name = sampled_df.loc[i, 'song']
        artists = sampled_df.loc[i, 'artist']
        print("{:<5} | {:<40} | {:<30} | {:.4f}".format(rank, song_name, artists, score))

# Get user input for song name and artist name
user_song = input("Enter a song name: ")
user_artist = input("Enter the artist name: ")

# Get similar songs using all three similarity metrics
get_similar_songs(user_song, user_artist, sampled_df, doc_embeddings)
```
